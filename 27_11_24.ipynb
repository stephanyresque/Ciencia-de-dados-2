```python
import pandas as pd

BASE = 'employee_attrition.csv'

df = pd.read_csv(BASE)
```

```python
df.head()
```

```python
with pd.option_context('display.max_columns', None):
    display(df.head())
```

```python
with pd.option_context('display.max_columns', None):
    display(df.info())
```

```python
with pd.option_context('display.max_columns', None):
    display(df.describe())
```

```python
colunas_valor_const = []

for coluna in df.columns:
  if df[coluna].nunique() == 1:
    colunas_valor_const.append(coluna)

colunas_valor_const
```

```python
df = df.drop(columns=colunas_valor_const, axis=1)

with pd.option_context('display.max_columns', None):
    display(df.head())
```

```python
df = df.drop(columns=['EmployeeNumber'], axis=1)

with pd.option_context('display.max_columns', None):
    display(df.head())
```

```python
colunas_categoricas_nao_ordenadas = [
    'BusinessTravel',
    'Department',
    'EducationField',
    'Gender',
    'JobRole',
    'MaritalStatus',
    'OverTime'
]
colunas_categoricas_ordenadas = [
    'Education',
    'EnvironmentSatisfaction',
    'JobInvolvement',
    'JobLevel',
    'JobSatisfaction',
    'PerformanceRating',
    'RelationshipSatisfaction',
    'StockOptionLevel',
    'WorkLifeBalance'
]
coluna_alvo = ['Attrition']

colunas_numericas = [
    coluna for coluna in df.columns
    if coluna not in (colunas_categoricas_nao_ordenadas + colunas_categoricas_ordenadas + coluna_alvo)
]
```

```python
colunas_numericas
```

```python
for coluna in (colunas_categoricas_ordenadas + colunas_categoricas_nao_ordenadas):
  print(f'{coluna}:{df[coluna].unique()}')
```

```python
for coluna in (colunas_categoricas_nao_ordenadas):
  df[coluna] = df[coluna].astype('category')


for coluna in (colunas_categoricas_ordenadas):
  df[coluna] = df[coluna].astype('category')

  categorias_ordenadas = sorted(df[coluna].unique())

  df[coluna] = df[coluna].astype('category').cat.set_categories(categorias_ordenadas, ordered=True)
```

```python
df.info()
```

```python
for coluna in (colunas_categoricas_ordenadas + colunas_categoricas_nao_ordenadas):
  print(f'{coluna}:{df[coluna].unique()}')
  print()
```
