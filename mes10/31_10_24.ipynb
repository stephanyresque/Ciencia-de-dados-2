```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
marcadores = np.where(vendas == np.min(vendas), 'x', '')

plt.plot(meses, vendas)

for posicao,marcador in enumerate(marcadores):
  if marcador == 'x':
    cor = 'red'
    tamanho = 80
  else:
    cor = 'blue'
    tamanho = 0
  plt.scatter(meses[posicao], vendas[posicao], marker = marcador, color = cor, s = tamanho)

plt.annotate('Menor valor', color = 'red', xy=(2,78), xytext = (2.3,78))

plt.xlabel('Meses')
plt.ylabel('Vendas (em milhares de dólares)')
plt.title('Crescimento das vendas no ano')
plt.show()
```

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
plt.bar(produtos, quantidades)
plt.show()
```

```python
barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels = quantidades, fontsize = 10, fontweight = 'bold', padding=3 )
plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])
plt.title('Quantidade de produtos vendidos')
plt.show()
```

```python
barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels = quantidades, fontsize = 10, fontweight = 'bold', padding=3 )
plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])

for i, barra in enumerate(barras):
  if i != quantidades.index(max(quantidades)):
    barras[i].set_color(cores[1])
    barras[i].set_alpha(0.3)

plt.title('Quantidade de produtos vendidos')
plt.show()
```

```python
rotulos = [valor if valor == max(quantidades) else '' for valor in quantidades]
```

```python
barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels = rotulos, fontsize = 10, fontweight = 'bold', padding=3 )
plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])

for i, barra in enumerate(barras):
  if i != quantidades.index(max(quantidades)):
    barras[i].set_color(cores[1])
    barras[i].set_alpha(0.3)

plt.title('Quantidade de produtos vendidos')
plt.show()
```

```python
rotulos = [valor if valor == max(quantidades_ordenadas) else '' for valor in quantidades_ordenadas]
```

```python
barras = plt.barh(produtos_ordenados, quantidades_ordenadas)

plt.bar_label(barras, labels = rotulos, fontsize = 10, fontweight = 'bold', padding=3 )
plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])

for i, barra in enumerate(barras):
  if i != quantidades_ordenadas.index(max(quantidades_ordenadas)):
    barras[i].set_color(cores[1])
    barras[i].set_alpha(0.3)

plt.title('Quantidade de produtos vendidos')
plt.show()
```

```python
barras = plt.barh(produtos_ordenados, quantidades_ordenadas)

plt.bar_label(barras, labels = rotulos, fontsize = 10, fontweight = 'bold', padding=3 )
plt.box(False)
plt.tick_params(axis='y', length=0)
plt.xticks([])

for i, barra in enumerate(barras):
  if i != quantidades_ordenadas.index(max(quantidades_ordenadas)):
    barras[i].set_color(cores[1])
    barras[i].set_alpha(0.3)

plt.title('Quantidade de produtos vendidos')
plt.gca().invert_yaxis()
plt.show()
```
