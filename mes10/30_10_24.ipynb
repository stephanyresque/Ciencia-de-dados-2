```python
import matplotlib.pyplot as plt
```

```python
from cycler import cycler
```

```python
ciclo_cores = cycler('color', plt.get_cmap('Accent').colors)
plt.rc('axes', prop_cycle = ciclo_cores)
```

```python
plt.scatter(ano,receita)
plt.xlabel('Anos')
plt.ylabel('Receita (em bilhões de dólares)')
plt.title('Receita anual Apple Inc')
plt.grid(True, linestyle = '--')
plt.show()
```

```python
fabricante = ['Samsung', 'Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Outros']
participacao = [30.2, 26.6, 10.4, 9.7, 4.8, 18.3]
destaque = [0.1,0.05,0,0,0,0]
```

```python
plt.figure(figsize = (8,8))
plt.pie(participacao, labels = fabricante, autopct = '%.1f%%', explode = destaque, shadow = True)
plt.title('Participação dos fabricantes no mercado')
plt.show()
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
plt.plot(meses, vendas, marker = 'x', markersize = 10, linestyle = '--', color = 'green', linewidth = 2)
plt.xlabel('Meses')
plt.ylabel('Vendas (em milhares de dólares)')
plt.title('Crescimento das vendas no ano')
plt.show()
```

```python
plt.plot(meses, vendas)

plt.annotate('Menor valor', color = 'red', xy=(2,78), xytext = (3,80), arrowprops=dict(facecolor = 'red'))

plt.xlabel('Meses')
plt.ylabel('Vendas (em milhares de dólares)')
plt.title('Crescimento das vendas no ano')
plt.show()
```

```python
import numpy as np
```

```python
marcadores = np.where(vendas == np.min(vendas), 'x', '')

plt.plot(meses, vendas)

for posicao,marcador in enumerate(marcadores):
  if marcador == 'x':
    cor = 'red'
    tamanho = 80
  else:
    cor = 'blue'
    tamanho = 0
  plt.scatter(meses[posicao], vendas[posicao], marker = marcador, color = cor, s = tamanho)

plt.annotate('Menor valor', color = 'red', xy=(2,78), xytext = (2.3,78))

plt.xlabel('Meses')
plt.ylabel('Vendas (em milhares de dólares)')
plt.title('Crescimento das vendas no ano')
plt.show()
```
