```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=meses, y=vendas);
```

```python
dados = {'Meses': meses, 'Vendas': vendas}

sns.lineplot(x = 'Meses', y = 'Vendas', data = dados)
```

```python
linha = sns.lineplot(x = 'Meses', y = 'Vendas', data = dados)

print(type(linha))
```

```python
linha = sns.lineplot(x = 'Meses', y = 'Vendas', data = dados);

linha.set_title('Vendas Mensais');
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (10, 6))

sns.lineplot(x = 'Meses', y = 'Vendas', data = dados, ax = ax)

ax.set_title('Vendas Mensais');

plt.show()
```

```python
cidades = ['Brasília', 'Rio de Janeiro', 'São Paulo', 'Curitiba', 'Manaus']
temperaturas = [27, 29, 25, 23, 31]

barras = sns.barplot(x = cidades, y = temperaturas);

barras.set_title('Temperaturas Médias');
barras.set_xlabel('Cidades');
barras.set_ylabel('Temperatura (ºC)');
```

```python
dados_2 = {'Cidades': cidades, 'Temperaturas': temperaturas}

barras = sns.barplot(x = 'Cidades', y = 'Temperaturas', data = dados_2);

barras.set_title('Temperaturas Médias');
```
