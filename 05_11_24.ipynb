```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# grfico 1 - linha ------------------------------------------------------------------
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

axs[0, 0].plot(meses, vendas)
axs[0, 0].set_xlabel('Ms')
axs[0, 0].set_ylabel('Vendas (em milhares de reais)')
axs[0, 0].set_title('Crescimento das Vendas ao Longo do Ano')

# grfico 2 - barras ------------------------------------------------------------------
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

axs[0, 1].bar(produtos, quantidades)
axs[0, 1].set_xlabel('Produto')
axs[0, 1].set_ylabel('Quantidade de vendas')
axs[0, 1].set_title('Quantidade de vendas por produto')

# grfico 3 - pizza ------------------------------------------------------------------
categorias = ['Eletrnicos', 'Roupas', 'Alimentos', 'Livros']
quantidades_vendidas = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

axs[1, 0].pie(quantidades_vendidas, labels=categorias, explode=destaque, autopct='%.1f')
axs[1, 0].set_title('Categorias de produtos vendidos')

# grfico 4 - barras empilhadas ------------------------------------------------------
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

axs[1, 1].bar(anos, vendas_online, label='Online')
axs[1, 1].bar(anos, vendas_loja, bottom=vendas_online, label='Loja')
axs[1, 1].set_xlabel('Ano')
axs[1, 1].set_ylabel('Vendas (em milhes de reais)')
axs[1, 1].set_title('Vendas online vs loja')
axs[1, 1].set_xticks(anos)
axs[1, 1].legend()

# ajustes finais ---------------------------------------------------------------------
fig.suptitle('Evoluo de vendas', fontsize=16)
plt.show()
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
novos_usuarios = [100, 80, 110, 140, 150, 160, 170, 120, 110, 150, 200, 140]
usuarios_ativos_diarios = [1000, 750, 900, 1300, 1400, 1500, 1200, 1200, 1500, 1900, 2000, 1800]
receita_mensal = [10000, 8500, 9800, 10000, 11000, 12000, 10000, 10500, 13000, 13000, 20000, 17000]
tickets_suporte_abertos = [110, 75, 100, 140, 170, 160, 100, 130, 190, 200, 210, 140]
tickets_suporte_fechados = [100, 85, 100, 120, 130, 150, 140, 110, 130, 120, 150, 200]
```

```python
mosaico = 'AA;BB;CD'
fig = plt.figure(figsize=(10,12))
espacamento = {'wspace':0.3, 'hspace':0.5}
axs = fig.subplot_mosaic(mosaico, gridspec_kw=espacamento)

#grfico 1
axs['A'].plot(meses, novos_usuarios)
axs['A'].set_ylabel('Novos usurios')
axs['A'].set_title('Novos usurios por ms')

#grfico 2
axs['B'].bar(meses, receita_mensal)
axs['B'].set_ylabel('Receita (em reais)')
axs['B'].set_title('Receita mensal')

#grfico 3
axs['C'].plot(meses, usuarios_ativos_diarios)
axs['C'].set_ylabel('Usurios ativos dirios')
axs['C'].set_title('Mdia de usurios ativos diariamente')

#grfico 4
axs['D'].bar(meses, tickets_suporte_abertos)
axs['D'].bar(meses, tickets_suporte_fechados, bottom=tickets_suporte_abertos)
axs['D'].set_ylabel('Tickets de suporte')
axs['D'].set_title('Tickets de suporte abertos e fechados')

#ajustes finais
fig.suptitle('Dados App X', fontsize= 16)
plt.show()
```
