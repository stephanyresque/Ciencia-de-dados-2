```python
import matplotlib.pyplot as plt
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
plt.plot(meses, vendas)
plt.show()
```

```python
plt.plot(meses, vendas, marker = 'o')
plt.show()
```

```python
plt.plot(meses, vendas, marker = 'o')
plt.xlabel('Meses')
plt.ylabel('Vendas (em milhares de reais)')
plt.title('Vendas durante o ano')
plt.show()
```

```python
plt.plot(meses, vendas, marker = 'o')
plt.xlabel('Meses')
plt.ylabel('Vendas (em milhares de reais)')
plt.title('Vendas durante o ano')
plt.grid(True)
plt.show()
```

```python
cidades = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte']
temperaturas = [25, 28, 22]
```

```python
plt.bar(cidades, temperaturas)
plt.show()
```

```python
plt.bar(cidades, temperaturas)
plt.ylabel('Temperatura média (°C)')
plt.title('Temperaturas nas Cidades')
plt.show()
```

```python
plt.bar(cidades, temperaturas, color = 'green')
plt.ylabel('Temperatura média (°C)')
plt.title('Temperaturas nas Cidades')
plt.show()
```

```python
barras = plt.bar(cidades, temperaturas, color = 'green')

plt.ylabel('Temperatura média (°C)')
plt.title('Temperaturas nas Cidades')
plt.show()
```

```python
barras = plt.bar(cidades, temperaturas, color = 'green')

plt.bar_label(barras, labels=temperaturas)

plt.ylabel('Temperatura média (°C)')
plt.title('Temperaturas nas Cidades')
plt.show()
```

```python
barras = plt.bar(cidades, temperaturas, color = 'green')

plt.bar_label(barras, labels=temperaturas, padding = 2, fontsize = 10, fontweight = 'bold')

plt.ylabel('Temperatura média (°C)')
plt.title('Temperaturas nas Cidades')
plt.show()
```

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
```

```python
plt.pie(quantidades)
plt.show()
```

```python
plt.pie(quantidades, labels = categorias)
plt.show()
```

```python
plt.pie(quantidades, labels = categorias)
plt.title('Distribuição percentual de categorias de produtos')
plt.show()
```

```python
plt.pie(quantidades, labels = categorias, autopct='%.1f')
plt.title('Distribuição percentual de categorias de produtos')
plt.show()
```

```python
destaques = [0.1,0,0,0]

plt.pie(quantidades, labels = categorias, autopct='%.1f', explode = destaques)
plt.title('Distribuição percentual de categorias de produtos')
plt.show()
```

```python
destaques = [0.1,0,0,0.1]

plt.pie(quantidades, labels = categorias, autopct='%.1f', explode = destaques)
plt.title('Distribuição percentual de categorias de produtos')
plt.show()
```

```python
destaques = [0.1,0.1,0,0]

plt.pie(quantidades, labels = categorias, autopct='%.1f', explode = destaques)
plt.title('Distribuição percentual de categorias de produtos')
plt.show()
```

```python
cores = plt.get_cmap('Pastel1').colors

plt.pie(quantidades, labels = categorias, autopct='%.1f', colors = cores)
plt.title('Distribuição percentual de categorias de produtos')
plt.show()
```

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]
```

```python
plt.bar(anos, vendas_online)
plt.show()
```

```python
plt.bar(anos, vendas_online)
plt.bar(anos, vendas_loja, bottom = vendas_online)
plt.show()
```

```python
plt.bar(anos, vendas_online, label = 'Online')
plt.bar(anos, vendas_loja, bottom = vendas_online, label = 'Loja')

plt.legend()

plt.show()
```

```python
plt.bar(anos, vendas_online, label = 'Online')
plt.bar(anos, vendas_loja, bottom = vendas_online, label = 'Loja')

plt.xticks(ticks=anos)

plt.legend()

plt.show()
```

```python
plt.bar(anos, vendas_online, label = 'Online')
plt.bar(anos, vendas_loja, bottom = vendas_online, label = 'Loja')

plt.xticks(ticks=anos, labels = ['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4'])

plt.legend()

plt.show()
```

```python
plt.bar(anos, vendas_online, label = 'Online')
plt.bar(anos, vendas_loja, bottom = vendas_online, label = 'Loja')

plt.xticks(ticks=anos)
plt.xlabel('Anos')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas de Produtos Online e Loja')

plt.legend()

plt.show()
```

```python
from cycler import cycler
```

```python
cores_cycler = cycler('color', plt.get_cmap('Pastel1').colors)
```

```python
plt.rc('axes', prop_cycle=cores_cycler) #configuração em tempo de execução
plt.bar(anos, vendas_online, label = 'Online')
plt.bar(anos, vendas_loja, bottom = vendas_online, label = 'Loja')

plt.xticks(ticks=anos)
plt.xlabel('Anos')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas de Produtos Online e Loja')

plt.legend()

plt.show()
```
