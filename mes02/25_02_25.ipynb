# Gradiente descendente

Como visto anteriormente, nosso objetivo  diminuir o erro de previso do nosso modelo. Para isso, precisamos ajustar os parmetros do modelo de forma que o erro seja minimizado. Para fazer isso, usamos o gradiente descendente. Mas antes de entender o gradiente descendente, precisamos entender o que  uma derivada.

## Derivada:

A derivada  uma ferramenta matemtica que nos ajuda a entender como uma funo muda quando seus valores de entrada mudam. Voc pode pensar na derivada como a "taxa de mudana" ou a "inclinao" de uma funo em um determinado ponto. 

Se voc tem uma funo $ f(x) $ que representa, por exemplo, o preo de uma casa baseado no tamanho da casa ($ x $), a derivada $ f'(x) $ lhe diria como o preo da casa muda quando o tamanho da casa muda.

A formula matemtica para a derivada de uma funo $ f(x) $ em relao a $ x $ :

$$ f'(x) = \lim_{{h \to 0}} \frac{{f(x + h) - f(x)}}{h} $$

Esta frmula pode parecer intimidante, mas a ideia principal  que estamos olhando para a mudana na funo $ f $ para uma pequena mudana em $ x $, e isso nos d a inclinao da funo em $ x $.

## Gradiente Descendente:

Agora, vamos avanar para o gradiente descendente. O gradiente descendente  um algoritmo usado para encontrar o ponto mnimo de uma funo, ou seja, onde a funo tem o valor mais baixo possvel.

Quando estamos construindo modelos de aprendizado de mquina, muitas vezes queremos minimizar uma funo de custo - uma medida de quo erradas esto nossas previses. O gradiente descendente nos ajuda a fazer isso ajustando os parmetros do modelo passo a passo para encontrar o ponto onde a funo de custo  mnima.

Aqui est como funciona o gradiente descendente:

1. Comeamos com valores aleatrios para os parmetros do nosso modelo.
2. Calculamos a derivada da funo de custo em relao a cada parmetro. Isso nos d a direo na qual a funo de custo est aumentando mais rapidamente.
3. Atualizamos os parmetros do modelo movendo-os na direo oposta  direo na qual a funo de custo est aumentando. Isso  feito subtraindo um pequeno valor (chamado taxa de aprendizado) multiplicado pela derivada dos parmetros.
  
A formula matemtica para a atualizao dos parmetros :

$$ \theta_{\text{novo}} = \theta_{\text{velho}} - \alpha \cdot \nabla J(\theta) $$

onde:
- $ \theta $ so os parmetros do modelo,
- $ \alpha $  a taxa de aprendizado,
- $ \nabla J(\theta) $  o gradiente da funo de custo $ J $ em relao aos parmetros $ \theta $.

Continuamos a atualizar os parmetros at que a mudana na funo de custo seja muito pequena, indicando que encontramos o mnimo.

A beleza do gradiente descendente  que ele usa a derivada para entender a direo na qual a funo de custo est aumentando, e ento ajusta os parmetros para mover-se na direo oposta para minimizar a funo de custo. Isso nos leva a um modelo bem ajustado que faz previses precisas.












## Exemplos

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
def gradiente_descendente(funcao_derivada, inicio, taxa_aprendizagem, n_iteracoes=50):
    """Funo que implementa o algoritmo de gradiente descendente de forma simplificada.

    Parameters
    ----------
    funcao_derivada : function
        Derivada da funo que se deseja minimizar.
    inicio : float
        Valor inicial para a busca.
    taxa_aprendizagem : float
        Valor do hiperparmetro de aprendizagem.
    n_iteracoes : int, opcional
        Nmero de iteraes, por padro 50

    Returns
    -------
    array
        Array com os valores de cada iterao.
    """
    vetor = [inicio]
    for _ in range(n_iteracoes):
        derivada = funcao_derivada(vetor[-1])
        atual = vetor[-1] - taxa_aprendizagem * derivada
        vetor.append(atual)
    return np.array(vetor)
```

```python
gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.1)
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.1)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.8)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.005)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.005, n_iteracoes=200)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = x**4 - 5 * x**2 - 3 * x

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(
    lambda x: 4 * x**3 - 10 * x - 3, inicio=0, taxa_aprendizagem=0.2
)

plt.plot(
    x_gradiente,
    x_gradiente**4 - 5 * x_gradiente**2 - 3 * x_gradiente,
    marker="o",
    c="red",
    linestyle="dashed",
)

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(
    lambda x: 4 * x**3 - 10 * x - 3, inicio=0, taxa_aprendizagem=0.1
)

plt.plot(
    x_gradiente,
    x_gradiente**4 - 5 * x_gradiente**2 - 3 * x_gradiente,
    marker="o",
    c="red",
    linestyle="dashed",
)

plt.grid()
plt.show()
```

# Regresso linear mltipla

```python
import pandas as pd

exemplo_03 = {
    "quartos": [1, 2, 3, 5, 6, 7],
    "preco": [155, 197, 244, 356, 407, 448],
    "idade": [10, 8, 7, 5, 3, 1]
}

df = pd.DataFrame(exemplo_03)

df
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(data = df, xs = 'quartos', ys='idade', zs='preco', c='red')

ax.set_xlabel('Quartos')
ax.set_ylabel('Idade')
ax.set_zlabel('Preo')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(df[['quartos', 'idade']], df['preco'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[4, 6]])
```

```python
y_pred = model.predict(df[['quartos', 'idade']])

y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

print(
    mean_absolute_error(df['preco'], y_pred),
    mean_squared_error(df['preco'], y_pred),
    root_mean_squared_error(df['preco'], y_pred),
    r2_score(df['preco'], y_pred)
)
```
