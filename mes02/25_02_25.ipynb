# Gradiente descendente

Como visto anteriormente, nosso objetivo é diminuir o erro de previsão do nosso modelo. Para isso, precisamos ajustar os parâmetros do modelo de forma que o erro seja minimizado. Para fazer isso, usamos o gradiente descendente. Mas antes de entender o gradiente descendente, precisamos entender o que é uma derivada.

## Derivada:

A derivada é uma ferramenta matemática que nos ajuda a entender como uma função muda quando seus valores de entrada mudam. Você pode pensar na derivada como a "taxa de mudança" ou a "inclinação" de uma função em um determinado ponto. 

Se você tem uma função $ f(x) $ que representa, por exemplo, o preço de uma casa baseado no tamanho da casa ($ x $), a derivada $ f'(x) $ lhe diria como o preço da casa muda quando o tamanho da casa muda.

A formula matemática para a derivada de uma função $ f(x) $ em relação a $ x $ é:

$$ f'(x) = \lim_{{h \to 0}} \frac{{f(x + h) - f(x)}}{h} $$

Esta fórmula pode parecer intimidante, mas a ideia principal é que estamos olhando para a mudança na função $ f $ para uma pequena mudança em $ x $, e isso nos dá a inclinação da função em $ x $.

## Gradiente Descendente:

Agora, vamos avançar para o gradiente descendente. O gradiente descendente é um algoritmo usado para encontrar o ponto mínimo de uma função, ou seja, onde a função tem o valor mais baixo possível.

Quando estamos construindo modelos de aprendizado de máquina, muitas vezes queremos minimizar uma função de custo - uma medida de quão erradas estão nossas previsões. O gradiente descendente nos ajuda a fazer isso ajustando os parâmetros do modelo passo a passo para encontrar o ponto onde a função de custo é mínima.

Aqui está como funciona o gradiente descendente:

1. Começamos com valores aleatórios para os parâmetros do nosso modelo.
2. Calculamos a derivada da função de custo em relação a cada parâmetro. Isso nos dá a direção na qual a função de custo está aumentando mais rapidamente.
3. Atualizamos os parâmetros do modelo movendo-os na direção oposta à direção na qual a função de custo está aumentando. Isso é feito subtraindo um pequeno valor (chamado taxa de aprendizado) multiplicado pela derivada dos parâmetros.
  
A formula matemática para a atualização dos parâmetros é:

$$ \theta_{\text{novo}} = \theta_{\text{velho}} - \alpha \cdot \nabla J(\theta) $$

onde:
- $ \theta $ são os parâmetros do modelo,
- $ \alpha $ é a taxa de aprendizado,
- $ \nabla J(\theta) $ é o gradiente da função de custo $ J $ em relação aos parâmetros $ \theta $.

Continuamos a atualizar os parâmetros até que a mudança na função de custo seja muito pequena, indicando que encontramos o mínimo.

A beleza do gradiente descendente é que ele usa a derivada para entender a direção na qual a função de custo está aumentando, e então ajusta os parâmetros para mover-se na direção oposta para minimizar a função de custo. Isso nos leva a um modelo bem ajustado que faz previsões precisas.












## Exemplos

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
def gradiente_descendente(funcao_derivada, inicio, taxa_aprendizagem, n_iteracoes=50):
    """Função que implementa o algoritmo de gradiente descendente de forma simplificada.

    Parameters
    ----------
    funcao_derivada : function
        Derivada da função que se deseja minimizar.
    inicio : float
        Valor inicial para a busca.
    taxa_aprendizagem : float
        Valor do hiperparâmetro de aprendizagem.
    n_iteracoes : int, opcional
        Número de iterações, por padrão 50

    Returns
    -------
    array
        Array com os valores de cada iteração.
    """
    vetor = [inicio]
    for _ in range(n_iteracoes):
        derivada = funcao_derivada(vetor[-1])
        atual = vetor[-1] - taxa_aprendizagem * derivada
        vetor.append(atual)
    return np.array(vetor)
```

```python
gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.1)
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.1)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.8)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.005)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(lambda x: 2 * x, inicio=5, taxa_aprendizagem=0.005, n_iteracoes=200)

plt.plot(x_gradiente, x_gradiente**2, marker="o", c="red", linestyle="dashed")

plt.grid()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = x**4 - 5 * x**2 - 3 * x

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(
    lambda x: 4 * x**3 - 10 * x - 3, inicio=0, taxa_aprendizagem=0.2
)

plt.plot(
    x_gradiente,
    x_gradiente**4 - 5 * x_gradiente**2 - 3 * x_gradiente,
    marker="o",
    c="red",
    linestyle="dashed",
)

plt.grid()
plt.show()
```

```python
plt.plot(x, y)

x_gradiente = gradiente_descendente(
    lambda x: 4 * x**3 - 10 * x - 3, inicio=0, taxa_aprendizagem=0.1
)

plt.plot(
    x_gradiente,
    x_gradiente**4 - 5 * x_gradiente**2 - 3 * x_gradiente,
    marker="o",
    c="red",
    linestyle="dashed",
)

plt.grid()
plt.show()
```

# Regressão linear múltipla

```python
import pandas as pd

exemplo_03 = {
    "quartos": [1, 2, 3, 5, 6, 7],
    "preco": [155, 197, 244, 356, 407, 448],
    "idade": [10, 8, 7, 5, 3, 1]
}

df = pd.DataFrame(exemplo_03)

df
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(data = df, xs = 'quartos', ys='idade', zs='preco', c='red')

ax.set_xlabel('Quartos')
ax.set_ylabel('Idade')
ax.set_zlabel('Preço')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(df[['quartos', 'idade']], df['preco'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[4, 6]])
```

```python
y_pred = model.predict(df[['quartos', 'idade']])

y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

print(
    mean_absolute_error(df['preco'], y_pred),
    mean_squared_error(df['preco'], y_pred),
    root_mean_squared_error(df['preco'], y_pred),
    r2_score(df['preco'], y_pred)
)
```
