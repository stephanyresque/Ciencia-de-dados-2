```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
BASE = 'top_10_ibov_mai_ago_2023.csv'

df = pd.read_csv(BASE)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
```

```python
df = df.set_index('Date', drop = True)
```

```python
df.head()
```

```python
df.index
```

```python
fig, ax = plt.subplots()


ax.plot(df.index, df['BBDC4'])

plt.show()
```

```python
fig, ax = plt.subplots()


ax.plot(df.index, df['BBDC4'], label = 'BBDC4')
ax.plot(df.index, df['ITUB4'], label = 'ITUB4')
ax.plot(df.index, df['BBAS3'], label = 'BBAS3')

ax.legend()
ax.set_title('Aes BBDC4, ITUB4 e BBAS3')
ax.set_ylabel('Valor em R$')

plt.show()
```

```python
variacao_diaria = df.pct_change()
variacao_diaria.head()
```

```python
fig, ax = plt.subplots()


ax.plot(variacao_diaria.index, variacao_diaria['BBDC4'], label = 'BBDC4')
ax.plot(variacao_diaria.index, variacao_diaria['ITUB4'], label = 'ITUB4')
ax.plot(variacao_diaria.index, variacao_diaria['BBAS3'], label = 'BBAS3')

ax.legend()
ax.set_title('Comparando a variao diria dos bancos top 10 IBOV')
ax.set_ylabel('Variao percentual')

plt.show()
```

```python
plt.style.use('bmh')
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize = (10, 10))

bancos = ['BBDC4', 'ITUB4', 'BBAS3']

axs[0].plot(variacao_diaria.index, variacao_diaria['BBDC4'], label = 'BBDC4', color = 'C0')
axs[0].legend()
axs[1].plot(variacao_diaria.index, variacao_diaria['ITUB4'], label = 'ITUB4', color = 'C1' )
axs[1].legend()
axs[2].plot(variacao_diaria.index, variacao_diaria['BBAS3'], label = 'BBAS3', color = 'C2')
axs[2].legend()

plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize = (10, 10))

bancos = ['BBDC4', 'ITUB4', 'BBAS3']

for i, banco in enumerate(bancos):
    axs[i].plot(variacao_diaria.index, variacao_diaria[banco], label = banco, color = 'C' + str(i))
    axs[i].legend()

plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize = (10, 10))

bancos = ['BBDC4', 'ITUB4', 'BBAS3']

for i, banco in enumerate(bancos):
    axs[i].plot(variacao_diaria.index, variacao_diaria[banco], label = banco, color = 'C' + str(i))
    axs[i].legend()

fig.suptitle('Comparando a variao diria dos bancos top 10 IBOV', fontsize = 16)
fig.supylabel('Variao percentual', fontsize = 16)

plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize = (10, 10), sharex = True)

bancos = ['BBDC4', 'ITUB4', 'BBAS3']

for i, banco in enumerate(bancos):
    axs[i].plot(variacao_diaria.index, variacao_diaria[banco], label = banco, color = 'C' + str(i))
    axs[i].legend()

fig.suptitle('Comparando a variao diria dos bancos top 10 IBOV', fontsize = 16)
fig.supylabel('Variao percentual', fontsize = 16)

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (10, 6))

axs[0].scatter('BBDC4','BBAS3', data = df)
axs[0].set_xlabel('BBDC4')
axs[0].set_ylabel('BBAS3')
axs[0].set_title('BBDC4 x BBAS3')

axs[1].scatter('BBDC4','ITUB4', data = df)
axs[1].set_xlabel('BBDC4')
axs[1].set_ylabel('ITUB4')
axs[1].set_title('BBDC4 x ITUB4')

axs[2].scatter('BBAS3','ITUB4', data = df)
axs[2].set_xlabel('BBAS3')
axs[2].set_ylabel('ITUB4')
axs[2].set_title('BBAS3 x ITUB4')

fig.suptitle('Analisando correlao dos bancos', fontsize = 16)

plt.show()
```

```python
from itertools import combinations

print(list(combinations(bancos, 2)))
```

```python
pares_bancos = list(combinations(bancos, 2))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize = (12, 8))

for i, par in enumerate(pares_bancos):
  axs[i].scatter(par[0], par[1], data = df)
  axs[i].set_xlabel(par[0])
  axs[i].set_ylabel(par[1])
  axs[i].set_title(f'{par[0]} x {par[1]}')

fig.suptitle('Analisando correlao dos bancos', fontsize = 16)

plt.show()
```
