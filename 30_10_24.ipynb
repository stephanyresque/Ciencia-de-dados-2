```python
import matplotlib.pyplot as plt
```

```python
ano = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
receita = [42.9, 65.2, 108.2, 156.5, 170.9, 182.8, 233.7, 215.6, 229.2, 265.6, 260.2, 274.5, 365.8, 394.3]
```

```python
from cycler import cycler
```

```python
ciclo_cores = cycler('color', plt.get_cmap('Accent').colors)
plt.rc('axes', prop_cycle = ciclo_cores)
```

```python
plt.scatter(ano,receita)
plt.xlabel('Anos')
plt.ylabel('Receita (em bilhões de dólares)')
plt.title('Receita anual Apple Inc')
plt.grid(True, linestyle = '--')
plt.show()
```

```python
plt.scatter(ano,receita, s = receita)
plt.xlabel('Anos')
plt.ylabel('Receita (em bilhões de dólares)')
plt.title('Receita anual Apple Inc')
plt.grid(True, linestyle = '--')
plt.show()
```

```python
plt.scatter(ano,receita, s = receita, c = receita,cmap = 'Reds')
plt.xlabel('Anos')
plt.ylabel('Receita (em bilhões de dólares)')
plt.title('Receita anual Apple Inc')
plt.grid(True, linestyle = '--')
plt.show()
```

```python
plt.scatter(ano,receita, s = receita, c = receita,cmap = 'Reds')
plt.xlabel('Anos')
plt.ylabel('Receita (em bilhões de dólares)')
plt.title('Receita anual Apple Inc')
plt.grid(True, linestyle = '--')

plt.colorbar()
plt.show()
```
