```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

fig, ax = plt.subplots()

ax.scatter(df.index, df['carrots'], label = 'observações')

ax.axhline(df['carrots'].mean(), color = 'C1', linestyle = '--', label = 'Média')
ax.axhline(df['carrots'].mean() + df['carrots'].std(), color = 'C2', linestyle = '--', label = 'Desvio Padrão')
ax.axhline(df['carrots'].mean() - df['carrots'].std(), color = 'C2', linestyle = '--', label = 'Desvio Padrão')

ax.xaxis.set_major_locator(mtick.MaxNLocator(integer = True))

ax.set_ylabel('Tempo (s) da última etapa da linha de produção')
ax.set_xlabel('Observações')
ax.legend()
plt.show()
```

```python
DATA = 'C:\\Users\\steph\\OneDrive\\Documentos\\MeusProjetos\\Ciencia-de-dados\\dados_hash\\plano_saude.csv'

df = pd.read_csv(DATA)

df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['operadora'] = df['operadora'].astype('category')

ordem = ['baixo', 'medio', 'alto']
df['satisfacao'] = pd.Categorical(df['satisfacao'], categories=ordem, ordered=True)

df.info()
```

```python
pd.crosstab(
    df['operadora'],
    df['satisfacao'],
    margins=True,
    margins_name='Total'
)
```

```python
pd.crosstab(
    df['operadora'],
    df['satisfacao'],
    margins=True,
    margins_name='Total',
    normalize='all'
)
```

```python
pd.crosstab(
    df['operadora'],
    df['satisfacao'],
    margins=True,
    margins_name='Total',
    normalize='all'
).style.format('{:.1%}')
```

```python
pd.crosstab(
    df['operadora'],
    df['satisfacao'],
    margins=True,
    margins_name='Total',
    normalize='index'
).style.format('{:.1%}')
```

```python
pd.crosstab(
    df['operadora'],
    df['satisfacao'],
    margins=True,
    margins_name='Total',
    normalize='columns'
).style.format('{:.1%}')
```

```python
pd.crosstab(
    df['operadora'],
    df['satisfacao'],
    margins=True,
    margins_name='Total'
)
```

```python
pd.crosstab(
    df['operadora'],
    df['satisfacao'],
    margins=True,
    margins_name='Total',
    normalize='index'
).style.format('{:.1%}')
```

```python
from scipy.stats import chi2_contingency
```

```python
resultado = chi2_contingency(
    pd.crosstab(df['operadora'], df['satisfacao'])
)

resultado
```

```python
resultado.statistic
```

```python
resultado.expected_freq
```

```python
esperados = resultado.expected_freq
```

```python
df_esperados = pd.DataFrame(
    esperados,
    index = df['operadora'].cat.categories,
    columns=df['satisfacao'].cat.categories
)

df_esperados
```

```python
resultado.pvalue
```
