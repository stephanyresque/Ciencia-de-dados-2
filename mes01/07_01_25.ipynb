```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng(seed = 42)

dados = rng.integers(1, 7, size = 100)

dados
```

```python
valores, contagem = np.unique(dados, return_counts=True)

print(valores, contagem)
```

```python
plt.vlines(valores, 0, contagem, linewidth=1, linestyle='dotted')

plt.scatter(valores, contagem, marker='o', color = 'red')

plt.show()
```

```python
def lancar_dados(vezes):
    rng = np.random.default_rng(seed = 42)
    dados = rng.integers(1, 7, size = vezes)

    valores, contagem = np.unique(dados, return_counts=True)
    print(valores, contagem)

    plt.vlines(valores, 0, contagem, linewidth=1, linestyle='dotted')
    plt.scatter(valores, contagem, marker='o', color = 'red')
    plt.show()
```

```python
lancar_dados(100)
```

```python
lancar_dados(1000)
```

```python
lancar_dados(10_000)
```

```python
lancar_dados(100_000)
```

```python
from scipy.stats import norm

normal = norm.rvs(0, 1, size=10_000, random_state = 42)
valores, contagem = np.unique(normal, return_counts=True)

print(valores)
print(contagem)

plt.scatter(valores, contagem, marker='o', color = 'red')

plt.show()
```

```python
x = np.linspace(-4, 4, 10_000)
y = norm.pdf(x, 0, 1)

plt.scatter(x, y, marker = 'o', color = 'red')
plt.plot(x, y)

plt.show()
```

```python
fig, ax = plt.subplots()

sns.histplot(normal, ax = ax, stat='density')

ax.plot(x, y)

plt.show()
```

```python
x = np.linspace(-4, 4, 10_000)
y = norm.cdf(x, 0, 1)

plt.plot(x, y)
plt.show()
```

```python
from scipy.stats import chi2

for grau_liberdade in range(1, 11):
    x = np.linspace(0, 20, 10_000)
    y = chi2.pdf(x, grau_liberdade)

    plt.plot(x, y, label = f'Graus de Liberdade: {grau_liberdade}')

plt.ylim(0,1)
plt.xlim(0,20)

plt.legend()
plt.show()
```
