```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='dark', palette='bright', context='notebook')

ARQUIVO_DADOS = 'churn_clientes.csv'

df = pd.read_csv(ARQUIVO_DADOS)
df = df.drop(columns='id_cliente', axis=1)


colunas_numericas = df.select_dtypes(include='number').columns
colunas_categoricas = df.select_dtypes(exclude='number').columns
colunas_categoricas = colunas_categoricas.drop('churn')

df.head()
```

```python
g = sns.FacetGrid(df, col = 'contrato')
g.map_dataframe(sns.histplot, x = 'meses_na_empresa');
```

```python
g = sns.FacetGrid(df, col = 'contrato')
g.map_dataframe(sns.histplot, x = 'meses_na_empresa')
plt.show()
```

```python
g = sns.FacetGrid(df, col = 'contrato', hue = 'churn')
g.map_dataframe(sns.histplot, x = 'meses_na_empresa')
plt.show()
```

```python
g = sns.FacetGrid(df, col = 'contrato', row = 'churn')
g.map_dataframe(sns.histplot, x = 'meses_na_empresa')
plt.show()
```

```python
g = sns.FacetGrid(df, col = 'contrato', row = 'churn')
g.map_dataframe(sns.kdeplot, x = 'meses_na_empresa')
plt.show()
```

```python
g = sns.FacetGrid(df, col = 'contrato', row = 'churn')
g.map_dataframe(sns.scatterplot, x = 'valor_mensal', y = 'total_gasto')
plt.show()
```

```python
g = sns.FacetGrid(df, col = 'contrato', hue = 'churn')
g.map_dataframe(sns.scatterplot, x = 'valor_mensal', y = 'total_gasto')
g.add_legend()
plt.show()
```
