```python
categorias = ['pessimo', 'ruim', 'regular', 'bom', 'otimo']

df_class['classificacao'] = df_class['classificacao'].astype('category')

df_class['classificacao'] = df_class['classificacao'].cat.set_categories(categorias, ordered=True)
```
