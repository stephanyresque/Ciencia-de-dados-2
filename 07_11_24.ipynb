```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

plt.style.use('fivethirtyeight')

params = {
    'axes.titlecolor':'dimgray',
    'figure.figsize':(12,6)
}

plt.rcParams.update(params)

barras = plt.barh(produtos, quantidades)

plt.bar_label(barras, labels=quantidades, padding=3, fontsize=10, fontweight='bold')

plt.xticks([])
plt.title('Quantidade de Produtos Vendidos')
plt.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(8, 4))

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

ax.plot(meses, vendas, marker='o')
ax.set_xlabel('Ms')
ax.set_ylabel('Vendas (em milhares)')
ax.set_title('Crescimento das Vendas ao Longo do Ano')
ax.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

params = {
    'axes.titlecolor':'dimgray',
    'figure.figsize':(12,6)
}

plt.rcParams.update(params)

fig, ax = plt.subplots()

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

ax.plot(meses, vendas, marker='o')
ax.set_xlabel('Ms')
ax.set_ylabel('Vendas (em milhares)')
ax.set_title('Crescimento das Vendas ao Longo do Ano')
ax.grid(False)

plt.show()
```
