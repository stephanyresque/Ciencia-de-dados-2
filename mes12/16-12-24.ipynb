```python
df_ordenado = df['carrots'].sort_values().reset_index(drop=True).to_frame()

df_ordenado
```

```python
df_ordenado['quartis'] = pd.qcut(df_ordenado['carrots'], q=4)

df_ordenado
```

```python
df_ordenado.info()
```

```python
df_ordenado['quartis'].cat.categories
```

```python
df_ordenado['quartis_rotulo'] = pd.qcut(
    df_ordenado['carrots'], 
    q=4,
    labels= ['0-25%', '25-50%', '50-75%', '75-100%']
)

df_ordenado
```
