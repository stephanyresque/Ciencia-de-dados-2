```python
import pandas as pd
```

```python
BASE = 'top_10_ibov_mai_ago_2023.csv'

df = pd.read_csv(BASE)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')

df.info()
```

```python
df.head()
```

```python
df = df.set_index('Date', drop = True)
```

```python
df.head()
```

```python
df.describe()
```

```python
 df.plot();
```

```python
import matplotlib.pyplot as plt

plt.style.use('bmh')
```

```python
df.plot()
plt.show()
```

```python
df.plot(figsize=(12, 9))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 9))

df.plot(ax=ax)

ax.set_title('Top 10 IBOV')

plt.show()
```
