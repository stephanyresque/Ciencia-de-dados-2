```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
fig, ax = plt.subplots()

ax.plot(meses, vendas)
ax.set_title('Vendas Mensais')
ax.set_xlabel('MÃªs')
ax.set_ylabel('Vendas')

plt.show()
```

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
fig, ax = plt.subplots()

ax.bar(produtos,quantidades)
ax.set_title('Quantidade de vendas por produto')
ax.set_xlabel('Produtos')
ax.set_ylabel('Quantidades')
plt.show()
```

```python
categorias = ['EletrÃ´nicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]
```

```python
fig, ax = plt.subplots()

ax.pie(quantidades, labels = categorias, explode = destaque, autopct="%.1f")

plt.show()
```

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]
```

```python
fig, ax = plt.subplots()

ax.bar(anos, vendas_online, label = 'Vendas Online')
ax.bar(anos, vendas_loja, bottom = vendas_online,label = 'Vendas Loja')

ax.legend()
ax.set_title('Vendas Online VS Loja')
ax.set_xlabel('Ano')
ax.set_ylabel('Vendas (em milhÃµes de reais)')

ax.set_xticks(anos)

plt.show()
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]
```

```python
fig, ax1 = plt.subplots()

cor_temperatura = 'tab:red'
cor_precipitacao = 'tab:blue'

ax1.bar(meses, precipitacao_media, color = cor_precipitacao, label = 'PrecipitaÃ§Ã£o mÃ©dia (mm)')
ax1.set_ylabel('PrecipitaÃ§Ã£o (mm)', color = cor_precipitacao)
ax1.tick_params(axis = 'y', labelcolor = cor_precipitacao)

ax2 = ax1.twinx()
ax2.plot(meses, temperatura_media, color = cor_temperatura, linewidth = 2, label = 'Temperatura mÃ©dia (CÂ°)')
ax2.set_ylabel('Temperatura (CÂ°)', color = cor_temperatura)
ax2.tick_params(axis = 'y', labelcolor = cor_temperatura)
ax2.set_ylim(20, 30)

ax1.set_title("Temperatura de precipitaÃ§Ã£o ao longo do ano")

fig.legend(bbox_to_anchor = (0.75,1), bbox_transform = ax1.transAxes)

plt.show()
```
