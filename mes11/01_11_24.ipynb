```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

# rc = runtime configuration (configuração quando executa)
cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

barras_online = plt.bar(anos, vendas_online, label='Online')
barras_fisica = plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja física')
plt.bar_label(barras_online, labels = vendas_online, label_type='center', fontweight = 'bold')
plt.bar_label(barras_fisica, labels = vendas_loja, label_type='center', fontweight = 'bold')

plt.tick_params(axis = 'x', length = 0)
plt.yticks([])
plt.box(False)
plt.xticks(ticks=anos)

plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```

```python
calorias_por_alimento = {
    "Batata Frita": 607,
    "Batata Chips": 542,
    "Bacon": 533,
    "Pizza": 296,
    "Cachorro Quente": 260
}
```

```python
alimentos = [alimento.replace(' ', '\n') for alimento in calorias_por_alimento.keys()]
```

```python
plt.figure(figsize=(8, 5))
barras = plt.bar(alimentos, calorias_por_alimento.values())

for i, barra in enumerate(barras):
  if list(calorias_por_alimento.keys())[i] == "Bacon":
        barra.set_color("red")
        barra.set_alpha(0.8)
  else:
      barra.set_color("gray")
      barra.set_alpha(0.5)

plt.bar_label(barras, fontweight = 'bold', color = 'white', padding = -15)
plt.tick_params(axis = 'x', length = 0)
plt.yticks([])
plt.box(False)

plt.title('Calorias por 100 g de alimento', color = 'gray', fontweight = 'bold', fontsize = 14)

plt.savefig('grafico_barras.png', bbox_inches = 'tight', dpi = 300)
plt.show()
```
