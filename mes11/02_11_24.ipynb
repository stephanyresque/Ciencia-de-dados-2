```python
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]
```

```python
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
cores = plt.get_cmap('Set3').colors
ciclo_cores = cycler('color', plt.get_cmap('Set3').colors)
plt.rc('axes', prop_cycle=ciclo_cores)

cor_anotacao = cores[4]
```

```python
plt.figure(figsize=(14,6))
plt.plot(anos, refeicoes, linewidth = 3, zorder = 1)

for posicao in (-1, 0):
  plt.scatter(anos[posicao], refeicoes[posicao], color = cor_anotacao)
  plt.annotate(refeicoes[posicao], color = cor_anotacao, xy = (anos[posicao], refeicoes[posicao]), xytext = (5,0), textcoords='offset points', size = 12, weight = 'bold')

plt.axvspan(2020,2021, alpha = 0.7, color = cores[1], zorder = 0)

plt.box(False)
plt.yticks([])

plt.xticks(anos)
plt.tick_params(axis = 'x', length = 0)
plt.text(2020, 40000, 'Pandemia', fontsize = 11, weight = 'bold', color = cor_anotacao)

plt.title('Refeições servidas por ano', fontsize = 16, weight = 'bold', loc = 'left', color = 'gray')

plt.savefig('refeicoes_por_ano.png')
plt.show()
```
