```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
fig, ax = plt.subplots()

ax.plot(meses, vendas)
ax.set_title('Vendas Mensais')
ax.set_xlabel('MÃªs')
ax.set_ylabel('Vendas')

plt.show()
```

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
fig, ax = plt.subplots()

ax.bar(produtos,quantidades)
ax.set_title('Quantidade de vendas por produto')
ax.set_xlabel('Produtos')
ax.set_ylabel('Quantidades')
plt.show()
```

```python
categorias = ['EletrÃ´nicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]
```

```python
fig, ax = plt.subplots()

ax.pie(quantidades, labels = categorias, explode = destaque, autopct="%.1f")

plt.show()
```

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]
```

```python
fig, ax = plt.subplots()

ax.bar(anos, vendas_online, label = 'Vendas Online')
ax.bar(anos, vendas_loja, bottom = vendas_online,label = 'Vendas Loja')

ax.legend()
ax.set_title('Vendas Online VS Loja')
ax.set_xlabel('Ano')
ax.set_ylabel('Vendas (em milhÃµes de reais)')

ax.set_xticks(anos)

plt.show()
```
